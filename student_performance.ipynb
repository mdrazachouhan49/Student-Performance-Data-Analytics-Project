```python
df1_cont = df.drop(['gender','race/ethnicity', 'parental level of education', 'lunch','test preparation course'], axis=1)
df1_cont.head()
```

```python
df1_cat = df.drop(['race/ethnicity', 'parental level of education', 'math score', 'reading score', 'writing score'], axis=1)
df1_cat.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()
MM_X = MM.fit_transform(df1_cont)
MM_X = pd.DataFrame(MM_X)
MM_X.columns = list(df1_cont)
MM_X.head().round(2)
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
LE_X = df1_cat.apply(LE.fit_transform)
LE_X.head()
```

```python
df1_final = pd.concat([MM_X, LE_X], axis=1)
df1_final.head()
```

```python
df1_final.corr()
```

```python
df1_final.corr()["math score"].sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df1_final.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
X = df1_final.drop(['math score'],axis=1)
```

```python
Y = df1_final['math score']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, random_state=44)
print(f'X trian size : {X_train.shape}')
print(f'Y trian size : {Y_train.shape}')
print(f'X test size : {X_test.shape}')
print(f'Y test size : {Y_test.shape}')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

MSE_A = mean_squared_error(Y_train, y_pred_train)
RMSE_A = np.sqrt(np.round(MSE_A,4))
R2_A = r2_score(Y_train, y_pred_train)

MSE_B = mean_squared_error(Y_test, y_pred_test)
RMSE_B = np.sqrt(np.round(MSE_B,4))
R2_B = r2_score(Y_test, y_pred_test)

print(f'Train MSE : {MSE_A:.4f}')
print(f'Train RMSE : {RMSE_A:.4f}')
print(f'Train r2 score : {R2_A:.2f}')

print(f'Test MSE : {MSE_B:.4f}')
print(f'Test RMSE : {RMSE_B:.4f}')
print(f'Test r2 score : {R2_B:.2f}')
```

```python
from sklearn.model_selection import ShuffleSplit, cross_validate

cv = ShuffleSplit(n_splits=500, test_size=0.20, random_state=42)
cv_score = cross_validate(model, X,Y, cv=cv, scoring='r2', return_train_score=True)

train_score = cv_score['train_score']
test_score = cv_score['test_score']

print(f'Cross Validation : Train Score : {train_score.mean():.2f}')
print(f'Cross Validation : Test Score : {test_score.mean():.2f}')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(Y_test, y_pred_test, alpha=0.6, label="Predicted vs Actual")
plt.plot(
    [Y_test.min(), Y_test.max()],
    [Y_test.min(), Y_test.max()],
    color='red', linestyle = '--', lw =2, label='Perfect fit (Y=X)')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.legend()
plt.tight_layout()
plt.show()
```
