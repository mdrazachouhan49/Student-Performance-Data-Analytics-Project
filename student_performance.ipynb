```python
X = df1[['reading score', 'writing score']]
```

```python
Y = df1['math score']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()


```

```python
model.fit(X,Y)
```

```python
y_pred = model.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
mse = mean_squared_error(Y,y_pred)
print("Mean squared error:", np.round(mse,2))
print("root Mean squared error:", np.round(np.sqrt(mse),2))
score = r2_score(Y,y_pred)
print("R square:", np.round(score,2))
VIF = 1/(1-score)
print("VIF :", np.round(VIF,2))
print("Model Accuracy :", np.round(score,2)*100,"%")
```
